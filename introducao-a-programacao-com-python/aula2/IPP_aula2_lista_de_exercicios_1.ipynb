##Funções

Uma função é um conjunto de instruções que possui um nome e que executa determinada tarefa. Já utilizamos várias funções nativas da linguagem Python como len, str, int e float. Agora, vamos criar as nossas funções.

Na próxima célula, é criada (ou definida) uma função. Mas ao executar a célula, nada acontece. Porque?

```python
# Criação de uma função
# Nome: volumeParalelepipedoV1
# Parâmetros: as 3 medidas de um paralelepípedo
# Valor de retorno: volume do paralelepípedo
#
def volumeParalelepipedoV1(comprimento, largura, altura):
    volume = comprimento * largura * altura
    return volume
```

Nada aconteceu porque a função só foi definida mas nunca utilizada em outra parte do programa.

Então, vamos utilizá-la. Como? Chamando-a pelo nome e fornecendo 3 valores conforme definido na linha de cabeçalho da função. Mas não se esqueça de executar a célula anterior para que o ambiente conheça a função *volumeParalelepipedo*.

```python
# Dois exemplos de uso da função volumeParalelepipedo
# A função é chamada duas vezes com argumentos diferentes
#
print('Exemplo 1: comprimento 10, largura 6 e altura 3')
vol = volumeParalelepipedoV1(10, 6, 3)
print(f'Volume: {vol}')

print('\nExemplo 2: comprimento 5.5, largura 4 e altura 1')
vol = volumeParalelepipedoV1(5.5, 4, 1)
print(f'Volume: {vol}')
```

Conforme a linha de cabeçalho da função, no momento da chamada da função, é necessário utilizar três argumentos. Se chamar a função com uma quantidade diferente de argumentos, ocorrerá um erro. Execute as duas próximas células e confira você mesmo.

Uma variável declarada no corpo de uma função só pode ser utilizada na própria função pois não existe fora dela. Seu escopo é local à função. Por isso, é possível utilizar variáveis com o mesmo nome em funções diferentes pois não haverá relação entre elas.

Execute a próxima célula e observe os valores impressos.

```python
# Funções
def f1():
    a = 10
    print(a)
    return

def f2():
    a = -9.9
    f1()
    print(a)
    return

# Bloco principal
f2()
```

Ao chamar a função *f2*, é criada a variável local *a* com o valor -9.9. Em seguida, o controle da execução passa para a função *f1* onde é criada uma outra variável *a*, local apenas à função em execução (*f1*). Em seguida, seu valor é impresso (10) e a função termina. O controle retorna para a função *f2*, onde é impresso o conteúdo da variável local *a* (-9.9) e a função encerra sua execução. O controle retorna para o bloco principal e o programa termina.

Na próxima célula, foi inserida a função *f3* que, ao ser executada, provocará a exibição de uma mensagem de erro pois não existe nenhuma variável a na função.

```python
# Funções
def f1():
    a = 10
    print(a)
    return

def f2():
    a = -9.9
    f1()
    print(a)
    return

def f3():
    print(a)
    return

# Bloco principal
f2()
a = 'variável definida no BP'
f3()
```

Agora, na célula anterior, retire o caractere de comentário ('#') na linha do bloco principal que cria a variável *a* com a string 'variável definida no BP' e execute novamente a célula.

Observe que o programa não acusa mais erro em sua execução porque a função f3 exibe o conteúdo da variável global *a*, definida no bloco principal.
